```
# 이미지 불러오기
img1 = cv2.imread('00039036.jpg')
img2 = cv2.imread('00080697.jpg')
```

```
# opencv로 이미지를 불러들이면 RGB가 아닌 BGR형태로 불러와지므로, 아래와 같이 색상이 좀 이상하게 보인다.
# 이미지는 0~255까지로 표현된 3차원 array 형태임, 흑백은 2차원(검/흰)
plt.imshow(img2)
```

```
# color 형태를 바꿔주면 제대로 보인다.
rgb_img = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```
# 이미지위에 사각형 그려보기 

size = 227
resized_img = cv2.resize(rgb_img, dsize=(size, size), interpolation=cv2.INTER_AREA)


# data_bbox 좌표 : 0.29, 0.11, 0.68, 0.89 (데이터 1x1 size 기준)
cv2.rectangle(resized_img, (int(0.29*size),int(0.11*size)), (int(0.68*size),int(0.89*size)), (255,0,0), 10)
plt.imshow(resized_img)
```

```
# data_eyes 좌표 : 0.41272727, 0.21429687
# 눈을 중심으로 얼굴의 위치좌표를 알 수 있음
cv2.circle(resized_img, (int(0.41272727*size),int(0.21429687*size)),2, (0,255,0),-1)  # 2 : 반지름 / -1 : 지정한 색으로 원채우기
plt.imshow(resized_img)
```

```
# data_gaze 좌표 : 0.51090909, 0.31132813
# 어딜보고 있는지 annotation이 되어 있음

cv2.circle(resized_img, (int(0.51090909*size),int(0.31132813*size)),2, (0,0,255),-1)  # -1 은 지정한 색으로 원채우기
plt.imshow(resized_img)
```
