```python
import networkx as nx
import matplotlib.pyplot as plt  # networkx 로 그림 그리려면 필요함
```

네트워크를 그리기 위해서는 점, 선, 점의 위치 정의필요

```python
G = nx.Graph()  # 빈 그래프 구조 G 생성
```

```python
G.add_edge(1,2)  # 엣지 추가 
```

```python
nx.draw(G, with_labels=True)  # 어떤 점이 1이고 2인지 알 수 있도록 표시해달라
plt.show()
```

```python
G.add_node('apple')  # apple 노드 추가 
```

```python
G.add_nodes_from(['banana', 'kiwi', 'mango'])  # 시리즈형태로 노드 추가 
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

```python
G = nx.MultiGraph()
G.add_nodes_from(['A','B','C','D'])
G.add_edges_from([('A','B'),('A','B'),('A','D'),('B','D'), ('B','C'), ('B','C'),('C','D')])
nx.draw(G, with_labels=True)
plt.show()
```

데이터는 캐글데이터셋 사용

```python
import pandas as pd
```

```python
data = pd.read_csv('data/Pokemon.csv')
poketmon = pd.DataFrame(data)
# stack_network_links.csv
# stack_network_nodes.csv'
```

## 포켓몬 타입별로 네트워크 그래프 그려보기 

```python
g = nx.Graph()
g = nx.from_pandas_edgelist(poketmon[:100], source='Name', target='Type 1')  # 100개의 관계도만 찍어보기 
print(nx.info(g))
```

```python
nx.draw(g)
plt.show(g)
```

```python
plt.figure(figsize=(20, 20))
pos = nx.spring_layout(g, k=0.15)  # position = pos
nx.draw_networkx(g,pos,node_size=25, node_color='red')
plt.show()
```

```python
g1 = nx.Graph()
g1 = nx.from_pandas_edgelist(poketmon[:100], source='Name', target='Type 2')  # 100개의 관계도만 찍어보기 
print(nx.info(g1))
```

```python
plt.figure(figsize=(20, 20))
pos = nx.spring_layout(g1, k=0.35)  # k 는 그래프자체가 크게 할 수록 벌어지는 것 같다. 
nx.draw_networkx(g1,pos,node_size=50, node_color='blue')
plt.show()
```

아 데이터셋 2개는 컬럼의 이름만 다르지 사실 같은 데이터를 가지고 있구나!

```python
plt.figure(figsize=(20, 20))
pos = nx.spring_layout(python, k=0.35)  # k 는 그래프자체가 크게 할 수록 벌어지는 것 같다. 
nx.draw_networkx(python,pos,node_size=50, node_color='blue')
plt.show()
```

## 그래프 좀 더 개선해보기 

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

df_nodes = pd.read_csv('data/stack_network_nodes.csv')
df_edges = pd.read_csv('data/stack_network_links.csv')

G = nx.Graph()

for index, row in df_nodes.iterrows():
    G.add_node(row['name'], group=row['group'], nodesize=row['nodesize'])
    
for index, row in df_edges.iterrows():
    G.add_weighted_edges_from([(row['source'], row['target'], row['value'])])
    
color_map = {1:'#f09494', 2:'#eebcbc', 3:'#72bbd0', 4:'#91f0a1', 5:'#629fff', 6:'#bcc2f2',  
             7:'#eebcbc', 8:'#f1f0c0', 9:'#d2ffe7', 10:'#caf3a6', 11:'#ffdf55', 12:'#ef77aa', 
             13:'#d6dcff', 14:'#d2f5f0'} 

plt.figure(figsize=(25,25))
options = {
    'edge_color': '#FFDEA2',
    'width': 1,
    'with_labels': True,
    'font_weight': 'regular',
}
colors = [color_map[G.node[node]['group']] for node in G]
sizes = [G.node[node]['nodesize']*10 for node in G]

nx.draw(G, node_color=colors, node_size=sizes, pos=nx.spring_layout(G, k=0.25, iterations=50), **options)
ax = plt.gca()
ax.collections[0].set_edgecolor("#555555") 
plt.show()
```
